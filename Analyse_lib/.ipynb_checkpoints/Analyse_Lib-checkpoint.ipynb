## Merge

```python
# Ventes et clients mergents avec grace au client_id
VC = pd.merge(Ventes , Clients)
VC
```

```python
#Check des infos si aucun valeur manquante
VC.info()
```

```python
# VC(Ventes, clients) et prod mergents avec grace au prod_id
PVC = pd.merge(Prod,VC)
PVC
```

```python
#Check des infos si aucun valeur manquante
PVC.info()
```

```python
PVC['price'].min()
```

```python
# Supprime les prix en dessous ou égale à 0 et même temps les les dates commencants par test.
PVC= PVC[PVC['price'] >= 0]
PVC.sort_values(by  = 'date')
```

```python
# Création du column Year_sell dans le quel il n'y a que les années.
Year_sell = []
for value in PVC["date"].values:# prends les dates
    Year_sell.append(int(value.split("-")[0]))#sépare les premiers valeurs avant le -.
    
PVC.insert(8, "Year_sell", np.array(Year_sell))
```

```python
PVC# Nouvelle colonne
```

```python
# Création de la colonne âge contenant les âges des clients lors d'achats
year = PVC['Year_sell'] # contient l'année d'achat
PVC['age'] = year - PVC['birth']# l'année d'achat - l'année de naissance
PVC
```

# Indice de Gini

```python
P = PVC[PVC['price'] > 0]
dep = P['price'].values
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0

xaxis = np.linspace(0-1/n,1+1/n,n+1)
plt.plot(xaxis,lorenz,drawstyle='steps-post')
plt.show()

AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
S = 0.5 - AUC # surface entre la première bissectrice et le courbe de Lorenz
gini = 2*S
gini
```

```python
B = PVC[PVC['birth'] >
            0]
dep = B['birth'].values
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0

xaxis = np.linspace(0-1/n,1+1/n,n+1) 
plt.plot(xaxis,lorenz,drawstyle='steps-post')
plt.title('')
plt.show()

AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
S = 0.5 - AUC # surface entre la première bissectrice et le courbe de Lorenz
gini = 2*S
gini
```

```python
YMP = PVC[['Year_sell' , 'Month', 'price']].groupby(['Year_sell','Month']).sum()
YMP.reset_index()
```
